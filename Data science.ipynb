```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv(r"C:\Users\Fariha\Downloads\archive\Iris.csv")

# 1. Count missing values
print("Missing values in each column:")
print(df.isnull().sum())

# 2. Show top 3 correlations (numeric only)
numeric_df = df.select_dtypes(include=['float64', 'int64'])
corr = numeric_df.corr().unstack().sort_values(ascending=False)

# Remove self-correlations (1.0) and duplicates
corr = corr[corr != 1].drop_duplicates().head(3)

print("\nTop 3 correlations:")
print(corr)

# 3. Plot one histogram (first numeric column)
if len(numeric_df.columns) > 0:
    col = numeric_df.columns[0]
    plt.hist(df[col].dropna(), bins=20, edgecolor='black')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
else:
    print("No numeric columns found for histogram.")
```
